```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_rows = None
df = pd.read_csv('raw_data.csv', index_col = 'Index')
```

* Steps to be taken

```python
# prasing salary estimate
# job description prasing
# prasing location
# prasing headquarter
# prasing founded
# prasing type of ownership
```

```python
# getting rid of those rows where salary estimate is missing
df = df[df['Salary Estimate'] != '-1']
print(df.shape)
df.head()
```

```python
# Cleaning salary estimate column
df['hourly'] = df['Salary Estimate'].apply(lambda x : 1 if 'Per Hour' in x else 0)
df['Provided_Salary'] = df['Salary Estimate'].apply(lambda x : 1 if 'Employer Provided Salary:' in x else 0)
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.split('(')[0])
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.replace('K', ''))
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.replace('$', ''))
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.replace('Employer Provided Salary:',''))
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.replace('Per Hour',''))
df['Min_Salary_Estimate'] = df['Salary Estimate'].apply(lambda x : x.split('-')[0])
df['Max_Salary_Estimate'] = df['Salary Estimate'].apply(lambda x : x.split('-')[1])
df[['Min_Salary_Estimate', 'Max_Salary_Estimate']] = df[['Min_Salary_Estimate', 'Max_Salary_Estimate']].astype(int)
df['Average_Salary_Estimate'] = (df['Min_Salary_Estimate'] + df['Max_Salary_Estimate'])/2
df.head()
```

```python
df[df['Rating'] < 0]
```

```python
# Company name 
df['company_txt'] = df.apply(lambda x : x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3], axis=1)
df['company_txt'] = df.apply(lambda x : x['company_txt'][:-3] if '\r\n' in x['company_txt'] else  x['company_txt'], axis=1)
```

```python
# Making new columns name job_states which captures the information of location of the job
df['job_state'] = df.apply(lambda x : x['Location'].split(',')[1], axis=1)

# Binary indicator for having same job location state where company headquarter is 
df['same_state'] = np.where(df['Location'] == df['Headquarters'], 1, 0)

# How old company is ?
df['company_age'] = np.where(df['Founded'] != -1, 2020-df['Founded'], df['Founded'])

# Visualisation
df[['job_state', 'same_state', 'company_age']]
```

```python
# Job Description
df['Job Description'][1]
```

```python
# Binary indicator to flag inportant tools in demand for data science career
def ds_tools(tool, column):
    df[tool] = df.apply(lambda x : 1 if tool in x[column].lower() else 0, axis=1 )
    print(f'Column : {tool} created')
    print(df[tool].value_counts(), '\n')

for var in ['python', 'excel', 'sas', 'spark', 'tableau']:
    ds_tools(var, 'Job Description')
```

```python
# Type of ownership
def ownership(company_type):
    list = company_type.split('-')
    if len(list) > 1:
        return list[1]
    else:
        return company_type
    
df['ownership'] = df['Type of ownership'].apply(ownership)
df['ownership']
```
