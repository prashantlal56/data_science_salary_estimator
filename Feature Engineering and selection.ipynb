```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import scipy.stats as stats
```

```python
# Configuring setting to display all the rows and columns
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
# Loading dataset
df=pd.read_csv('cleaned_data.csv')
df.head()
```

# Selecting columns

```python
data = df[['Rating', 'Industry', 'Sector', 'hourly', 'Revenue', 'Size', 'Provided_Salary',
           'Average_Salary_Estimate', 'company_txt', 'job_state', 'same_state', 'company_age',
           'python', 'excel', 'sas', 'spark', 'tableau', 'aws', 'big data', 'R_lang', 'ownership',
           'job_title', 'seniority', 'competitor_counts']]

data.head()
```

```python
print('Rating column having missing value :',len(data[data['Rating']==-1])/len(data)*100,'%')
print('company_age column having missing value :',len(data[data['company_age']==-1])/len(data)*100,'%')
```

```python
data['Rating'] = np.where(data['Rating']==-1, 0, data['Rating'])
data['company_age'] = np.where(data['company_age']==-1, 0, data['company_age'])

data.describe()
```

#### Making a dataset with dummies and saving it

```python
df_dummy = pd.get_dummies(data, drop_first=True)
df_dummy.to_csv('df_dummy.csv', index=False)
df_dummy.shape
```

# Spliting into train and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('Average_Salary_Estimate', axis=1),
                                                    data['Average_Salary_Estimate'],
                                                         test_size=0.15,
                                                         random_state=101)
                                                         
X_train.shape, X_test.shape
```

```python
num_var = [var for var in X_train.columns if X_train[var].dtypes != 'object']
cat_var = [var for var in X_train.columns if X_train[var].dtypes == 'object']

num_var, cat_var
```

# Encoding Rare labels

```python
# Encoding less frequent labels with 'Rare'
from feature_engine.categorical_encoders import RareLabelCategoricalEncoder

rare_encoder = RareLabelCategoricalEncoder(tol=0.03,n_categories=4, variables=cat_var, replace_with='Rare')
rare_encoder.fit(X_train)

X_train = rare_encoder.transform(X_train)
X_test = rare_encoder.transform(X_test)
```

# Integer encoding of categorical variables

```python
# Integer encoding of categorical variables
from feature_engine.categorical_encoders import OrdinalCategoricalEncoder

ordinal_encoder = OrdinalCategoricalEncoder(encoding_method='ordered', variables=cat_var)

ordinal_encoder.fit(X_train, y_train)

X_train = ordinal_encoder.transform(X_train)
X_test = ordinal_encoder.transform(X_test)
```

# Yeo-Johnson transform of continuous variables

```python
cont_var = [var for var in X_train.columns if X_train[var].nunique() > 20]

discrete_var = [var for var in num_var if var not in cont_var]

cont_var, discrete_var
```

```python
def cont_analyz(df, column):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    df[column].hist(bins=30)
    plt.title(column)
    
    plt.subplot(1,2,2)
    stats.probplot(df[column], dist='norm', plot=plt)
    plt.show()
    
for var in cont_var:
    cont_analyz(X_train, var)
```

```python
from feature_engine.variable_transformers import YeoJohnsonTransformer

def cont_analyz_power(df, list):
    exp_trans = YeoJohnsonTransformer(variables = list)
    exp_trans.fit(df)
    df=exp_trans.transform(df)
    for column in list:
        plt.figure(figsize=(12,5))
        plt.subplot(1,2,1)
        df[column].hist(bins=30)
        plt.title(column)
    
        plt.subplot(1,2,2)
        stats.probplot(df[column], dist='norm', plot=plt)
        plt.show()
        
cont_analyz_power(X_train, ['company_age', 'Rating'])
```

```python
# Yeo jonshon transform
yeo_johnson_transform = YeoJohnsonTransformer(variables = cont_var)
yeo_johnson_transform.fit(X_train)

X_train = yeo_johnson_transform.transform(X_train)
X_test = yeo_johnson_transform.transform(X_test)
```

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
```
